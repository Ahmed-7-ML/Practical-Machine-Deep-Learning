## *We will talk about `Exploratory Data Analysis (EDA)` using Pandas& Matplotlib/Seabon*

```python
import pandas as pd     # For data manipulation
import numpy as np      # For numerical operations
import matplotlib.pyplot as plt     # For plotting
import seaborn as sns       # For advanced plotting

# To display plots inline in Jupyter notebooks
%matplotlib inline
```

```python
df = pd.read_csv('iris.csv', )  # Load your csv file into a DataFrame

df.head()  # Display the first 5 rows of the DataFrame
```

```python
print(df.shape)  # Print the shape of the DataFrame (rows, columns)

print(df.info())  # Get a concise summary of the DataFrame
```

```python
# Drop unused Columns
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df.describe()  # Get statistical summary of numerical columns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['sepal length (cm)'], kde=True)
plt.title('Distribution of Your Sepal Length Column')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
for col in df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(x=col, data=df, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
df.columns
```

```python
# BIVARIATE ANALYSIS
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=df)
plt.title('Relationship Between Two Numerical Columns')
plt.show()
```

```python
for col in df.columns:
    if df[col].dtype in [np.float64, np.int64] and col != 'sepal length (cm)':
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x='sepal length (cm)', y=col, data=df)
        plt.title(f'Relationship Between Sepal Length and {col}')
        plt.xlabel('sepal length (cm)')
        plt.ylabel(col)
        plt.show()
```

```python
# CORRELATION MATRIX
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
