### _`Types of Plots in Seaborn :`_

- Relational Plots -> Figure Level = sns.relplot()

### _`Line Plot`_


```python
sns.lineplot(data=flights, x='year', y='passengers')
plt.title('Number of Passengers per Year')
plt.show()
```

```python
sns.lineplot(data=flights, x='year', y='passengers')
plt.title('Number of Passengers per Year')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

#### _`Scatter Plot`_

- Discuss the relationship between 2 numerical variables.


```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.title('Total Bill vs Tip (Per Gender)')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', style='smoker', size='size')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
sns.set_style('darkgrid')
```

```python
# Manually set the size range
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', style='smoker', size='size',
                sizes=(20, 200), alpha=0.7)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', palette='cool')
plt.title('Total Bill vs Tip (Per Gender)')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='time', style='time',
                markers={'Lunch': 'o', 'Dinner': 'D'})
plt.title('Total Bill vs Tip (Per Time)')
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips,
            col='sex', hue='time', style='sex',
            kind='scatter')
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips,
            row='day', col='sex', hue='time', style='sex',
            kind='scatter')
plt.show()
```

### _`Regression Plots`_


```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.title('Total Bill vs. Tip & the Best Fit Line ')
plt.show()
```

### _`Distribution Plots`_

- `Histogram`
- `KDE` = `Kernel Density Estimate`

  - _visualizing the probability density of a continuous variable. It provides a smoothed representation of the underlying distribution of a dataset. useful when dealing with continuous data or when you want to explore the distribution without making assumptions about a specific parametric form._

- `ECDF` = `Empirical Cumulative Distribution Function`
  _statistical tool used to visualize the distribution of a dataset. It represents the proportion or count of observations falling below each unique value in a dataset. Unlike histograms or density plots, ECDFs do not require binning or smoothing parameters, making them a straightforward way to visualize data._

- `RUG` :
  _plot of data for a single quantitative variable, displayed as marks along an axis. visualise the distribution of the data. As such it is analogous to a histogram with zero-width bins, or a one-dimensional scatter plot. often used in combination with two-dimensional scatter plots by placing a rug plot of the x values of the data along the x-axis, and similarly for the y values._


```python
df = sns.load_dataset('penguins')

df.head()
```

_`Histogram`_


```python
sns.histplot(data=df, x='bill_length_mm')
plt.show()
```

```python
sns.histplot(data=df, x='bill_length_mm', bins=20)
plt.show()
```

```python
sns.histplot(data=df, x='bill_length_mm', kde=True)
plt.show()
```

```python
sns.histplot(data=df, x='bill_length_mm', stat='density')
plt.show()
```

```python
sns.histplot(data=df, x='bill_length_mm', hue='species')
plt.show()
```

```python
sns.histplot(data=df, x='bill_length_mm', hue='species', multiple='stack')
plt.show()
```

```python
sns.histplot(data=df, x='bill_length_mm',
             hue='species', multiple='stack',
             element='step')
plt.show()
```

```python
sns.histplot(data=df, x='bill_length_mm',
             stat='density',
             hue='species', multiple='stack')
plt.show()
```

```python
sns.histplot(data=df, x='bill_length_mm',
             stat='density',
             hue='species', multiple='stack',
             common_norm=True)
plt.show()
```

```python
sns.histplot(data=df, x='species', stat='density')
plt.show()
```

```python
sns.histplot(data=df, x='species', stat='density', shrink=0.9)
plt.show()
```

```python
# Horizontal Histogram
sns.histplot(data=df, y='species', stat='density', shrink=0.9)
plt.show()
```

_`KDE Plot`_


```python
sns.kdeplot(data=df, x='bill_length_mm')
plt.show()
```

_`ECDF Plot`_


```python
sns.ecdfplot(data=df, x='bill_length_mm')
plt.show()
```

_`RUG Plot`_


```python
sns.rugplot(data=df, x='bill_length_mm')
plt.show()
```
